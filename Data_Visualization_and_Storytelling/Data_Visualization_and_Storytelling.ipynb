```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("graduation_rate.csv")

df = pd.DataFrame(data)

print(df.head())

# Step 2: Data Preparation

edu_order = ['high school', 'some college', "associate's degree", "bachelor's degree", "master's degree", "doctoral degree"]

# Check for missing values
print(df.isnull().sum())
```

#  3: Create Visualizations

# 3a. College GPA by Parental Education (Bar plot)

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="parental level of education", y="college gpa", data=df, order=edu_order, palette="viridis", errorbar=None)
plt.title("Average College GPA by Parental Education", fontsize=16)
plt.xlabel("Parental Education")
plt.ylabel("College GPA")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 3b. High School GPA vs College GPA (Scatter plot)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="high school gpa", y="college gpa", hue="years to graduate", size="parental income", sizes=(50, 200), data=df, palette="coolwarm")
plt.title("HS GPA vs College GPA (Colored by Years to Grad)", fontsize=16)
plt.xlabel("High School GPA")
plt.ylabel("College GPA")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# 3c. Parental Income Distribution by Years to Graduate (Box plot)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="years to graduate", y="parental income", data=df, palette="Set2")
plt.title("Parental Income Distribution by Years to Graduate", fontsize=16)
plt.xlabel("Years to Graduate")
plt.ylabel("Parental Income ($)")
plt.tight_layout()
plt.show()
```

# 3d. Heatmap of Correlation between Numerical Features

```python
plt.figure(figsize=(10, 6))
numeric_cols = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_cols.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap Between Features", fontsize=16)
plt.tight_layout()
plt.show()
```

#  4: Combine Visualizations for Storytelling

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 7))

# Plot 1: Bar Plot (Education vs GPA)
sns.barplot(x="parental level of education", y="college gpa", data=df,
            order=edu_order, palette="viridis", errorbar=None, ax=axes[0, 0])
axes[0, 0].set_title("Avg College GPA by Parent Education", fontsize=12)
axes[0, 0].tick_params(axis='x', rotation=45)

# Plot 2: Scatter Plot (HS GPA vs College GPA)
sns.scatterplot(x="high school gpa", y="college gpa", hue="years to graduate",
                data=df, palette="coolwarm", s=100, ax=axes[0, 1])
axes[0, 1].set_title("HS GPA vs College GPA", fontsize=12)

# Plot 3: Box Plot (Income vs Years to Grad)
sns.boxplot(x="years to graduate", y="parental income", data=df, palette="Set2", ax=axes[1, 0])
axes[1, 0].set_title("Income vs Years to Graduate", fontsize=12)

# Plot 4: Correlation Heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, ax=axes[1, 1])
axes[1, 1].set_title("Correlation Heatmap", fontsize=12)

plt.tight_layout()
plt.show()

# Step 5: Key Insights
insights = {
    "Education Impact": "Students with parents holding higher degrees (Bachelor's/Master's) tend to have higher College GPAs.",
    "Performance Trend": "High School GPA is a strong predictor of College GPA. Students who performed well in HS continued to do so in college.",
    "Economic Impact": "There is a visible trend where students taking longer to graduate (6-7 years) come from lower income households, possibly due to financial stress.",
    "Correlation Insights": "ACT and SAT scores are highly correlated (>0.90), meaning they provide redundant information. HS GPA correlates positively with College GPA."
}

for key, value in insights.items():
    print(f"{key}: {value}\n")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


data = pd.read_csv("graduation_rate.csv")
df = pd.DataFrame(data)

print(df.head())

edu_order = ['high school', 'some college', "associate's degree", "bachelor's degree", "master's degree", "doctoral degree"]

print(df.isnull().sum())
```

#  3: Create Visualizations

# 3a. Pairplot to show relationships between features


```python
subset_df = df[['high school gpa', 'college gpa', 'parental income', 'years to graduate']]
sns.pairplot(subset_df, hue="years to graduate", palette="viridis")
plt.suptitle("Pairplot: Relationships Between GPAs and Income", y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```

# 3b. Boxplot for College GPA distribution by Parental Education

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="parental level of education", y="college gpa", data=df, order=edu_order, palette="Set3")
plt.title("College GPA Distribution by Parental Education", fontsize=16)
plt.xlabel("Parental Education")
plt.ylabel("College GPA")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 3c. Violin plot for SAT Score distribution by Years to Graduate

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x="years to graduate", y="SAT total score", data=df, palette="Pastel1")
plt.title("SAT Score Distribution by Years to Graduate", fontsize=16)
plt.xlabel("Years to Graduate")
plt.ylabel("SAT Total Score")
plt.tight_layout()
plt.show()
```

# 3d. Heatmap of Correlation between features

```python
plt.figure(figsize=(10, 6))
numeric_cols = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_cols.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap Between Features", fontsize=16)
plt.tight_layout()
plt.show()
```

#  4: Combine the Visualizations for Storytelling



```python
fig, axes = plt.subplots(2, 2, figsize=(15, 7))

# 1. Scatter Plot (Top-Left) - Replaces Pairplot
sns.scatterplot(x="high school gpa", y="college gpa", hue="years to graduate",
                size="parental income", data=df, palette="viridis", ax=axes[0, 0])
axes[0, 0].set_title("High School GPA vs College GPA")

# 2. Boxplot (Top-Right)
sns.boxplot(x="parental level of education", y="college gpa", data=df,
            order=edu_order, palette="Set3", ax=axes[0, 1])
axes[0, 1].set_title("College GPA by Parent Education")
axes[0, 1].tick_params(axis='x', rotation=45)

# 3. Violin plot (Bottom-Left)
sns.violinplot(x="years to graduate", y="SAT total score", data=df, palette="Pastel1", ax=axes[1, 0])
axes[1, 0].set_title("SAT Scores by Years to Graduate")

# 4. Heatmap (Bottom-Right)
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, ax=axes[1, 1])
axes[1, 1].set_title("Correlation Heatmap")

plt.tight_layout()
plt.show()

# Step 5: Key Insights
insights = {
    "Overall Trend": "Students with higher previous academic performance (HS GPA/SAT) tend to graduate faster and maintain higher college GPAs.",
    "Pairplot/Scatter": "There is a clear positive relationship between High School GPA and College GPA. Students who take longer to graduate (hue) generally cluster in the lower GPA regions.",
    "Education Distribution": "The boxplot suggests that students whose parents have advanced degrees (Master's/PhD) tend to have a higher median College GPA.",
    "SAT & Graduation": "The violin plot shows that students graduating in 3-4 years typically have higher SAT scores compared to those taking 6-7 years.",
    "Correlation Insights": "SAT and ACT scores are highly correlated (>0.90). Parental Income also shows a positive correlation with College GPA."
}

for key, value in insights.items():
    print(f"{key}: {value}\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("graduation_rate.csv")

df = pd.DataFrame(data)

edu_order = ['high school', 'some college', "associate's degree", "bachelor's degree", "master's degree", "doctoral degree"]
```

# Step 3: Create Visualizations

# 3a. SAT vs College GPA (Scatter Plot with Trend)



```python
plt.figure(figsize=(10, 6))
sns.regplot(x='SAT total score', y='college gpa', data=df, color='b', ci=None)
plt.title("Trend: SAT Scores vs College GPA", fontsize=16)
plt.xlabel("SAT Total Score")
plt.ylabel("College GPA")
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```


# 3b. Average Income by Parental Education (Bar plot)

```python
plt.figure(figsize=(10, 6))
edu_income = df.groupby('parental level of education')['parental income'].mean().reindex(edu_order).dropna()
sns.barplot(x=edu_income.index, y=edu_income.values, palette='Set2')
plt.title("Average Parental Income by Education Level", fontsize=16)
plt.xlabel("Parental Education")
plt.ylabel("Average Income ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```


# 3c. Count of Students by Years to Graduate (Bar plot)

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='years to graduate', data=df, palette='viridis')
plt.title("Student Distribution by Years to Graduate", fontsize=16)
plt.xlabel("Years to Graduate")
plt.ylabel("Number of Students")
plt.tight_layout()
plt.show()
```

# 3d. ACT vs SAT Correlation (Line/Scatter plot)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='ACT composite score', y='SAT total score', data=df, s=100, color='g', marker='o')
plt.title("Correlation: ACT vs SAT Scores", fontsize=16)
plt.xlabel("ACT Score")
plt.ylabel("SAT Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```

#  4: Combine the Visualizations for Storytelling

```python

fig, axes = plt.subplots(2, 2, figsize=(15, 7))

# Plot 1: Trend (SAT vs College GPA)
sns.regplot(x='SAT total score', y='college gpa', data=df, color='b', ci=None, ax=axes[0, 0])
axes[0, 0].set_title("Trend: SAT vs College GPA")

# Plot 2: Bar (Income by Education)
sns.barplot(x=edu_income.index, y=edu_income.values, palette='Set2', ax=axes[0, 1])
axes[0, 1].set_title("Avg Income by Parent Education")
axes[0, 1].set_ylabel("Income ($)")
axes[0, 1].tick_params(axis='x', rotation=45)

# Plot 3: Count (Years to Graduate)
sns.countplot(x='years to graduate', data=df, palette='viridis', ax=axes[1, 0])
axes[1, 0].set_title("Count of Students by Years to Grad")
axes[1, 0].set_ylabel("Count")

# Plot 4: Scatter (ACT vs SAT)
sns.scatterplot(x='ACT composite score', y='SAT total score', data=df, s=100, color='g', marker='o', ax=axes[1, 1])
axes[1, 1].set_title("Correlation: ACT vs SAT")
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()

# Step 5: Key Insights
insights = {
    "Overall Trend": "There is a positive linear trend between SAT scores and College GPA; students with higher entrance scores tend to perform better in college.",
    "Income & Education": "Parental income scales distinctly with education level. The 'Doctoral' and 'Master's' categories show the highest average household income.",
    "Graduation Timeline": "The majority of students in this sample graduate within 5 years, though there is a long tail of students taking 6 or 7 years.",
    "Test Correlation": "ACT and SAT scores show a near-perfect linear correlation. A student scoring high on one almost certainly scores high on the other."
}

print("\n--- KEY INSIGHTS ---")
for key, value in insights.items():
    print(f"{key}: {value}\n")
```
